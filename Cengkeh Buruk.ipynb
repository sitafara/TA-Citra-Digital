```python
ukuran_baik = 0
warna_baik = 0

for index,value in enumerate(hist_h[1]):
    if (value >= 0.01 and value <= 0.07) and (hist_s[1][index] >= 0.1 and hist_s[1][index] <= 0.6):
        ukuran_baik += 1
    
    if (value >= 0.6 and value <= 1) and (hist_s[1][index] >= 0 and hist_s[1][index] <= 0.15):
        warna_baik += 1
ukuran_baik = ukuran_baik/hist_h[1].shape[0]*100
#print(ukuran_baik)
#print(warna_baik)

ukuran = False
warna = False
if ukuran_baik>0.01:
    ukuran = True
    
if warna_baik<500:
    warna = True
    
if ukuran==True and warna==True:
    print("Cengkeh Baik")
else:
    print("Cengkeh Buruk")
```
